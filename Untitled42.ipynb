```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC  # Example classifier (SVM)
from sklearn.model_selection import train_test_split

# Parameters for EEG signal
fs = 256  # Sampling frequency (Hz)
duration = 5  # Duration of the signal (seconds)
t = np.linspace(0, duration, int(fs * duration), endpoint=False)  # Time vector
frequency = 10  # Alpha wave frequency (Hz)
amplitude = 100  # Amplitude of the wave

# Generate alpha wave (synthetic EEG signal)
eeg_signal = amplitude * np.sin(2 * np.pi * frequency * t)

# Add Gaussian noise to simulate real EEG signal noise
noise = np.random.normal(0, 20, len(t))
eeg_signal_noisy = eeg_signal + noise

# Preprocessing: Apply a simple bandpass filter to remove noise
from scipy.signal import butter, filtfilt

def bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, data)

# Apply bandpass filter to remove noise
filtered_signal = bandpass_filter(eeg_signal_noisy, 8, 12, fs)  # Alpha range

# Feature Extraction: Frequency Spectrum (Power Spectrum)
freqs = np.fft.rfftfreq(len(t), 1/fs)
fft_vals = np.abs(np.fft.rfft(filtered_signal))

# Plot Time-domain Signals
plt.figure(figsize=(12, 6))
plt.subplot(3, 1, 1)
plt.plot(t, eeg_signal, label='Original Signal')
plt.title("Original EEG Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.subplot(3, 1, 2)
plt.plot(t, eeg_signal_noisy, label='Noisy Signal', color='orange')
plt.title("Noisy EEG Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.subplot(3, 1, 3)
plt.plot(t, filtered_signal, label='Filtered Signal', color='green')
plt.title("Filtered EEG Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()

# Plot Frequency Spectrum
plt.figure(figsize=(10, 4))
plt.plot(freqs, fft_vals, label='Frequency Spectrum')
plt.title("Frequency Spectrum of Filtered Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.show()

# Simulate more features for classification (generate 100 random values for demo purposes)
# Normally, you'd extract more features from your actual EEG signal.
features = np.random.rand(100)  # Simulating 100 feature values (e.g., average spectral power)
labels = np.random.randint(0, 2, 100)  # Random labels (0 or 1) for simplicity

# Split into training/testing for simulation
X_train, X_test, y_train, y_test = train_test_split(features.reshape(-1, 1), labels, test_size=0.2)

# Train the classifier (SVM)
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)

# Predict the class for the test set
predicted = svm.predict(X_test)
print(f"Predicted classes: {predicted}")

# Real-Time Control Simulation: Map signal intensity to action
for prediction in predicted:
    if prediction == 1:
        print("Action: Move cursor")
    else:
        print("Action: No movement")
```
